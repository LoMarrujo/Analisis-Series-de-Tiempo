```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

np.random.seed(1)
```

```python
#  1. white noise
##  sample size = 100 (or length of series)
T = 100

##  distribution with cero meand and finite variance
white_noise_rv = uniform(loc=-0.25, scale=0.5)

##  mean and var
white_noise_rv_mean, white_noise_rv_var = white_noise_rv.stats(moments='mv')

##  sample
white_noise_sample = white_noise_rv.rvs(size=T)
```

```python
##  mean and var

white_noise_rv_mean, white_noise_rv_var
```

```python
##  sample mean and var
white_noise_sample.mean(), white_noise_sample.var()
```

```python
#  uniform white noise plot
fig, ax= plt.subplots(figsize=(19,6))
plt.plot(white_noise_sample, color='k', label='Trajectory')
plt.hlines(y=0, xmin=-1, xmax=T+1,
    linestyle=':', color='red', label='Mean')

plt.title('Trajectory of Uniform White Noise')
plt.xlabel(r'$t$')
plt.ylabel(r'$X_t$')
plt.legend()
ax.set_xlim(-0.5, T+0.5)
```
